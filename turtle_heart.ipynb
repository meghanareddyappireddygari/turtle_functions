```python
import ColabTurtle.Turtle as tim
from random import randint

def initializeTimeTheTurtle():
  initializeTurtle()
  penup()
  goto(400, 10)
  pendown()
  tim.right(90)

def drawShapes(num_sides):
  tim.color(randint(0,220),randint(0, 220),randint(0, 220))
  for _ in range(num_sides):
    angle = 360/num_sides
    tim.forward(100)
    tim.right(angle)
```

```python
def curve():
  for i in range(200):
    tim.right(1)
    tim.forward(1)

def heart():

  tim.speed(10)


  tim.left(140)
  tim.forward(113)

  curve()
  tim.left(120)

  curve()
  tim.forward(112)
 
```
